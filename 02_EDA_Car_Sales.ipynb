## EDA - 02 - Car Sale Dataset

**About Dataset :**

The dataset provided focuses on used car sales, capturing various attributes to analyze the second-hand car market.

This dataset provides insights into factors affecting resale value, trends in the used car industry, and consumer behavior. 

**Source :**

The dataset is available on Kaggle, representing data commonly seen in online car resale platforms.  

Here is the dataset link :  https://drive.google.com/file/d/1N0kM_zi-iU3yvr6BEJf21aiF599PtA0U/view

**Description:**

The dataset contains details about used cars listed for sale, such as brand, model, selling price, kilometers driven, fuel type, and transmission type.This information is valuable for predictive modeling, market analysis, and understanding customer preferences. 

**Features:**

- Car_id: A unique identifier for each car in the dataset, helping to track individual car entries.

- Date: The date when the car sale transaction took place, formatted as YYYY-MM-DD.

- Customer Name: The name of the customer who purchased the car, represented as a string.

- Gender: The gender of the customer, categorized as "Male" or "Female."

- Annual Income: The customer's annual income in US dollars, represented as a numeric value.

- Dealer_Name: The name of the dealership selling the car, represented as a string.

- Company: The manufacturer or brand name of the car, such as "Toyota," "Ford," etc.

- Model: The specific model name of the car, such as "Corolla," "Civic," etc.

- Engine: The engine type of the car, such as "V6," "I4," etc.

- Transmission: The type of transmission in the car, either "Manual" or "Automatic.

- Color: The color of the car, represented as a string (e.g., "Red," "Blue").

- Price ($): The selling price of the car in US dollars.

- Dealer_No: A unique identifier for each car dealer in the dataset.

- Body Style: The body style of the car, such as "Sedan," "SUV," etc.

- Phone: The phone number of the customer who purchased the car.

- Dealer_Region: The geographical region of the car dealer, such as "North," "South," etc.

### Q.1 What is the average selling price of cars for each dealer, and how does it compare across different dealers ?

```python
df['Price'] = df['Price ($)'].astype(float)
```

```python
df.drop('Price ($)', axis=1, inplace=True)
```

```python
# What is the average selling price of cars for each dealer?
avg_sell_price_dealer = df.groupby('Dealer_Name')['Price'].mean().sort_values(ascending=False).round(2)
```

```python
avg_sell_price_dealer
```

```python
# how does it compare across different dealers ?

plt.figure(figsize=(14,6))
sns.barplot(x=avg_sell_price_dealer.index, y=avg_sell_price_dealer.values,palette = "bright")
plt.xticks(rotation=90)
plt.title('Average Selling Price of Cars for each Dealer')
plt.xlabel('Dealer Name')
plt.ylabel('Average Selling Price')
plt.show()
```

```python
# Insight: Dealer Named U-Haul co  has the highest average selling price of 28769.92 $ cars
```

### Q.2  Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends ?

```python
#  Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing strategy of the trends ?

car_brand_price = df.groupby('Company')['Price'].std().sort_values(ascending=False).round(2)

high_var_company = car_brand_price.idxmax()
high_var_price = car_brand_price.max()

print(f'The car brand with the highest variation in prices is : {high_var_company}')
print(f'The highest variation in prices is : {high_var_price} $')
```

```python
car_brand_price # Variation in prices of different car brands
```

```python
# Visualizing the variation in prices of cars for each company
plt.figure(figsize=(12,6))
sns.barplot(x=car_brand_price.index, y=car_brand_price.values,palette='bright')

plt.xticks(rotation=45)
plt.title('Variation in Prices of Cars for each Company')
plt.xlabel('Company')
plt.ylabel('Price Variation')
plt.tight_layout()
plt.show()
```

```python
# what does this tell us about the pricing strategy of the trends ?

# Insight: The car brand named 'Lincoin' has the highest variation in prices.
# Brands with a High Standard Deviation in Price : Likely offer a wide range of models targeting different customer segments.
# These companies aim to dominate multiple market niches. 
# High variation might also indicate frequent promotions, customizations, or premium add-ons.
```

### Q.3 What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare ?

```python
# What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare ?

plt.figure(figsize=(12,8))
sns.boxplot(x='Transmission', y='Price', data=df, palette='bright')
plt.title('Distribution of Car Prices for each Transmission Type')

plt.show()
```

```python
df.groupby('Transmission')['Price'].describe()
```

```python
iqr_comparison = df.groupby('Transmission')['Price'].quantile([0.25, 0.75]).unstack()
iqr_comparison['IQR'] = iqr_comparison[0.75] - iqr_comparison[0.25]
```

```python
iqr_comparison
```

```python
sns.barplot(x=iqr_comparison.index, y=iqr_comparison['IQR'], palette='bright')
plt.title('Interquartile Range Comparison for each Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Interquartile Range')
plt.show()
```

```python
# insight: The interquartile range for manual transmission cars is higher than automatic transmission cars.
```

### Q.10 What are the top 5 car models with the highest number of sales, and how does their price distribution look ?
