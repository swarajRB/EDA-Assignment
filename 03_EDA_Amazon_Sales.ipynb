## 03 - EDA - Amazon Sales Dataset

**Description:**

This dataset contains information on 1K+ Amazon products, including their ratings, reviews, and other details. 

**Features:** 
- product_id: Unique identifier for each product 

- product_name: Name of the product 
 
- category: Category of the product 

- discounted_price: Discounted price of the product 

- actual_price: Actual price of the product 

- discount_percentage: Percentage of discount for the product 

- rating: Rating of the product (1-5) 

- rating_count: Number of people who voted for the Amazon rating 

- about_product: Description about the product 

- user_id: ID of the user who wrote the review 

- user_name: Name of the user who wrote the review 

- review_id: ID of the user review 

- review_title: Short review 

- review_content: Long review 

- img_link: Image link of the product 

- product_link: Official website link of the product

## Data Cleaning and Preparation

```python
# Dropping some columns which are not necessary to keep

columns_to_drop = [
    'product_id',  
    'user_id',     
    'user_name',  
    'review_id',  
    'img_link',   
    'product_link',
    'about_product'
]

df = df.drop(columns=columns_to_drop)
```

```python
df
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.discounted_price.unique()
```

```python
df['discounted_price'] = df['discounted_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
```

```python
df.actual_price.unique()
```

```python
df['actual_price'] = df['actual_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
```

```python
df.rating.unique()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df.dtypes
```

```python
df.discount_percentage.unique()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)
```

```python
# Convert 'category' and 'product_name' to category type
df['category'] = df['category'].astype('category')
df['product_name'] = df['product_name'].astype('category')
```

```python
# Drop rows where 'rating_count' is NaN
df = df.dropna(subset=['rating_count'])
df['rating_count'] = df['rating_count'].replace({',': ''}, regex=True).astype(int)
```

```python
df.dtypes
```

### Q.1 What is the average rating for each product category ?

```python
avg_rating_by_category = df.groupby('category')['rating'].mean().reset_index()
avg_rating_by_category
```

```python
df['main_category'] = df['category'].str.split('|').str[0]

average_rating_by_main_category = df.groupby('main_category')['rating'].mean().reset_index()

# Display the average rating for each main category
average_rating_by_main_category
```

```python
plt.figure(figsize=(12, 8))  # Adjust the figure size

sns.barplot(x = "main_category",y = "rating",data = average_rating_by_main_category , palette="coolwarm")

# Add labels and title
plt.xlabel('Product Category')
plt.ylabel('Average Rating')
plt.title('Average Rating for Each Product Category')
plt.xticks(rotation = 90)
# Show the plot
plt.tight_layout()  # To adjust for label visibility
plt.show()
```

### Q.2 What are the top rating_count products by category ?

```python
data = df[['product_name','category','rating_count']]

top_rating_by_category = data.sort_values(by=['category','rating_count'],ascending=[True,False])

top_rating_by_category.groupby('category').head(1)
```

### Q.3 What is the distribution of discounted prices vs. actual prices ?

```python
# Set up the plot
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(x='discounted_price', y='actual_price', data=df, alpha=0.6, color = 'red')

# Add labels and title
plt.xlabel('Discounted Price (₹)')
plt.ylabel('Actual Price (₹)')
plt.title('Scatter Plot: Discounted Price vs Actual Price')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 10))

sns.pairplot(df[['discounted_price', 'actual_price']])

# Add title
plt.suptitle('Pairplot: Discounted Price vs Actual Price', y=1.02)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.histplot(df['discounted_price'], kde=True, label='Discounted Price', color='blue', alpha=0.6)
sns.histplot(df['actual_price'], kde=True, label='Actual Price', color='orange', alpha=0.6)

# Add labels and title
plt.xlabel('Price (₹)')
plt.ylabel('Frequency')
plt.title('Distribution of Discounted Prices vs Actual Prices')
plt.legend()

# Show the plot
plt.show()
```

### Q.4 How does the average discount percentage vary across categories ?

```python
df
```

```python
# Extract the main category
df['main_category'] = df['category'].str.split('|').str[0]



# Group by main category and calculate the average discount percentage
avg_discount_by_main_category = df.groupby('main_category')['discount_percentage'].mean().reset_index()
avg_discount_by_main_category
```

### Q.5 What are the most popular product names ?

### Q.6 What are the most popular product keywords ?

```python
from collections import Counter
import pandas as pd

df_cleaned = df.dropna(subset=['product_name'])  

all_words = ' '.join(df_cleaned['product_name']).split()

# Count the frequency of each word
word_counts = Counter(all_words)

# Convert the counts into a DataFrame for better readability
word_df = pd.DataFrame(word_counts.items(), columns=['Word', 'Count'])

sorted_words = word_df.sort_values(by='Count', ascending=False)

# Display the top 10 most popular words
top_words = sorted_words.head(20)
top_words
```

### Q.7 What are the most popular product reviews ?

```python
from collections import Counter
import pandas as pd

df_cleaned = df.dropna(subset=['review_content'])  

all_words = ' '.join(df_cleaned['review_content']).split()

# Count the frequency of each word
word_counts = Counter(all_words)

# Convert the counts into a DataFrame for better readability
word_df = pd.DataFrame(word_counts.items(), columns=['Word', 'Count'])

sorted_words = word_df.sort_values(by='Count', ascending=False)

# Display the top 10 most popular words
top_words = sorted_words.head(20)
top_words
```

### Q.8 What is the correlation between discounted_price and rating ?

```python
# Calculate the correlation between discounted_price and rating
corr_disprice_rating = df[['discounted_price', 'rating']].corr()

corr_disprice_rating
```

```python
# interpretation :
# The weak positive correlation suggests that the discounted price does not have much influence on the rating of products in the dataset.
#  Other factors likely play a more significant role in determining the ratings.
```
