## 04 - EDA - "Spotify Data: Popular Hip-hop Artists and Tracks"

**Description of the Dataset:** 

The dataset titled "Spotify Data: Popular Hip-hop Artists and Tracks" provides a curated collection of approximately 500 entries showcasing the vibrant realm of hip-hop music. 
These entries meticulously compile the most celebrated hip-hop tracks and artists, reflecting their significant influence on the genre's landscape. 
Each entry not only highlights the popularity and musical composition of the tracks but also underscores the creative prowess of the artists and their profound impact on global listeners. 


**Application in Data Science:** 

This dataset serves as a valuable resource for various data science explorations. Analysts can delve into trend analysis to discern the popularity dynamics of hit hip-hop tracks over recent years.

Additionally, the dataset enables network analysis to uncover collaborative patterns among top artists, shedding light on the genre's evolving collaborative landscape. 
Furthermore, it facilitates the development of predictive models aimed at forecasting track popularity based on diverse features, offering insights for artists, producers, and marketers. 


**Column Descriptors:**  

- Artist: The name of the artist, providing direct attribution to the creative mind behind the track. 

- Track Name: The title of the track, encapsulating its identity and essence. 

- Popularity: A numeric score reflecting the track's reception and appeal among Spotify listeners. 

- Duration (ms): The track's length in milliseconds, detailing the temporal extent of the musical experience. 

- Track ID: A unique identifier within Spotify's ecosystem, enabling direct access to the track for further exploration.

### Q.2 What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram ?

```python
# Plotting the histogram for 'Popularity'

plt.figure(figsize=(10, 6))
plt.hist(df['Popularity'], bins=20, color='pink', edgecolor='black')

plt.title('Distribution of Popularity Among Tracks', fontsize=16)
plt.xlabel('Popularity', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Q.3 Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Duration (ms)', y='Popularity', color='blue', alpha=0.6, edgecolor='black')

plt.title('Relationship Between Duration and Popularity of Tracks', fontsize=16)
plt.xlabel('Duration (ms)', fontsize=12)
plt.ylabel('Popularity', fontsize=12)
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
correlation = df['Duration (ms)'].corr(df['Popularity'])
print(f'Correlation coefficient between Duration and Popularity: {correlation:.2f}')
```

```python
# There is weak relation between the Duration and Popularity of Tracks
```

### Q.4 Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.

```python
# Counting the number of tracks for each artist
artist_counts = df['Artist'].value_counts().reset_index()

top_artists = artist_counts.head(10)
top_artists
```

```python

# Plotting the countplot for the top 10 artists
plt.figure(figsize=(12, 6))
sns.barplot(x="Artist", y="count",data = top_artists, palette='bright')

plt.title('Top 10 Artists with the Most Tracks', fontsize=16)
plt.xlabel('Artist', fontsize=12)
plt.ylabel('Number of Tracks', fontsize=12)
plt.xticks(rotation=90, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### Q.5 What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each ?

```python
# Sorting the dataset by Popularity in ascending order
least_popular_tracks = df.sort_values(by='Popularity', ascending=True).head(5)

print("Top 5 Least Popular Tracks:")
top_5_least_popular = least_popular_tracks[['Artist', 'Track Name','Popularity']]
top_5_least_popular
```

### Q.6 Among the top 5 most popular artists, which artist has the highest popularity on average? Calculate and display the average popularity for each artist ?

```python

artist_popularity = df.groupby('Artist')['Popularity'].sum()

# top 5 most popular artists
top_5_artists = artist_popularity.nlargest(5)

top_5_artist_tracks = df[df['Artist'].isin(top_5_artists.index)]

# average popularity for each of the top 5 artists
average_popularity = top_5_artist_tracks.groupby('Artist')['Popularity'].mean().reset_index()

average_popularity
```

### Q.8 Visualize relationships between multiple numerical variables simultaneously using a pair plot

```python
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Creating a pair plot
sns.pairplot(df[numerical_columns], kind='scatter', diag_kind='hist', palette='bright')

# Adding title to the plot
plt.suptitle('Pair Plot of Numerical Variables', fontsize=16, y=1.02)
plt.show()
```

### Q.9 Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot ?

```python
plt.figure(figsize=(19, 12))
sns.boxplot(data=df, x='Artist', y='Duration (ms)', palette='bright')

# Adding title and labels
plt.title('Distribution of Track Duration Across Different Artists', fontsize=16)
plt.xlabel('Artist', fontsize=12)
plt.ylabel('Track Duration (ms)', fontsize=12)
plt.xticks(rotation=90, ha='right')  # Rotate artist names for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

### Q.10 #10  How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot

```python
plt.figure(figsize=(19, 6))
sns.swarmplot(data=df, x='Artist', y='Popularity', palette='coolwarm')

# Adding title and labels
plt.title('Distribution of Track Popularity Across Different Artists', fontsize=16)
plt.xlabel('Artist', fontsize=12)
plt.ylabel('Track Popularity', fontsize=12)
plt.xticks(rotation=90)  # Rotate artist names for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
